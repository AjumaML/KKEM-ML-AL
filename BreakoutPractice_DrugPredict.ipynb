```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#reading data
drug = pd.read_csv('/content/drug200.csv')
```

```python
drug.head()
```

```python
#Encoding using Onehot

drug = pd.get_dummies(data = drug,columns = ['Sex','BP','Cholesterol'])
```

```python
drug.head()
```

```python
#excluding drug column creating X
X = drug.drop(['Drug'], axis=1)
```

```python
X.head()
```

```python
#creating Y
Y = drug['Drug']
```

```python
Y.head()
```

```python
# training and testing data splitting
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.75, random_state = 0)
```

# Logistic Regression

```python
#Creating Regression Model

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.linear_model import LogisticRegression
LM = LogisticRegression(max_iter=5000)
```

```python
#fitting Model
LM.fit(X_train,Y_train)
```

```python
#Predict data
Y_pred = LM.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_pred,Y_test)
```

# **KNN**

```python
#importing KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
#creating model 
KN = KNeighborsClassifier(n_neighbors=4)
```

```python
#fitting model
KN.fit(X_train, Y_train)
```

```python
#Prediction

Y_pred = KN.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_pred,Y_test)
```

```python
#predicting using values form 1 to 20 and finding max accuracy 
for i in range(1,21):
  KN = KNeighborsClassifier(n_neighbors=i)
  KN.fit(X_train,Y_train)
  Y_pred = KN.predict(X_test)
  acc = accuracy_score(Y_test, Y_pred)
  print('for n_neighbors%2d accuracy is %f'%(i,acc))
```

Maximum Accuracy = 0.64 for n_neighbors = 4
