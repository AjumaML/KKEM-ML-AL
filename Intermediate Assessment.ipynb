# Modelling

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
```

```python

from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test,Y_pred)
```

```python
from sklearn.neighbors import KNeighborsRegressor

KN = KNeighborsRegressor(n_neighbors=10)
KN.fit(X_train, Y_train)
Y_pred_KN = KN.predict(X_test)

mean_squared_error(Y_test, Y_pred_KN)

```

```python
from sklearn.ensemble import RandomForestRegressor

RF = RandomForestRegressor(random_state=45,max_depth=16,n_estimators=200)
RF.fit(X_train, Y_train)
Y_pred_RF = RF.predict(X_test)

mean_squared_error(Y_test, Y_pred_RF)
#print('{:0.2f}'.format(mse_rf))
```

After modeeling with three models, LinearRegression, KNeighborsRegressor and RandomForestRegressor; RandomForestRegressor gives the least mse - 1155505.4921311522

So, going forward with this model and applying fine tuning.

Fine Tuning with Hyperpot

```python
#best possible RandomForestRegressor model using hyperopt

from hyperopt import tpe, hp, fmin, STATUS_OK,Trials
```

```python
space ={'n_estimators':hp.uniform('n_estimators',100,500),
           'max_depth':hp.uniform('max_depth',5,20),
           'min_samples_leaf':hp.uniform('min_samples_leaf',1,5),
           'min_samples_split':hp.uniform('min_samples_split',2,6)}
```

```python
# define objective function

from sklearn.model_selection import cross_val_score

def hyperparameter_tuning(params):
    clf = RandomForestRegressor(**params,n_jobs=-1)
    acc = mean_squared_error(Y_test, Y_pred_RF).mean()
    return {"loss": -acc, "status": STATUS_OK}
```

```python
# Initialize trials object
trials = Trials()

best = fmin(
    fn=hyperparameter_tuning,
    space = space, 
    algo=tpe.suggest, 
    max_evals=50, 
    trials=trials
)

print("Best: {}".format(best))
```

# Best Fit after Hyperparameter Tuning

```python

RF = RandomForestRegressor(random_state=45, max_depth=7, n_estimators=189, min_samples_split=2, min_samples_leaf=3)
RF.fit(X_train, Y_train)
Y_pred_RF = RF.predict(X_test)

mse_RF = mean_squared_error(Y_test, Y_pred_RF)
print('{:0.2f}'.format(mse_RF))
```
