```python
#import necessary libraries
import numpy as np
import pandas as pd
```

```python
#reading data
df = pd.read_csv('/content/diabetes.csv')
```

```python
df.info()
```

```python
df.columns
```

```python
#Splitting into features and labels
X = df.drop('Outcome', axis=1)
y = df.Outcome
```

### Finding number of zeros in each coloumn

```python
for column_name in df.columns:
  count = (df[column_name] == 0).sum()
  print(column_name,count)
```

### Replacing zeroes with mean of each coloumn

```python
for column_name in X.columns:
  column = X[column_name]
  column.replace(to_replace = 0, value = column.mean(), inplace=True)
```

```python
X.head()
```

```python
y.head()
```

```python
#Test data and training data formation
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```


# Prediction using different models
### 1.Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
D_clf = DecisionTreeClassifier()
D_clf = D_clf.fit(X_train,y_train)
```

```python
y_pred = D_clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

### 2. Bagging Classifier with Decision Tree

```python
from sklearn.ensemble import BaggingClassifier
tree = DecisionTreeClassifier()
```

```python
bagging_clf = BaggingClassifier(base_estimator=tree, n_estimators=100, random_state = 42)
```

```python
bagging_clf.fit(X_train,y_train)
```

```python
y_pred = bagging_clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

## 3. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
  
RF_clf = RandomForestClassifier(n_estimators = 100) 
  
RF_clf.fit(X_train, y_train)

y_pred = RF_clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

### 4. Bagging Classifier with Random Forest

```python
Rtree = RandomForestClassifier()
```

```python
Rbagging_clf = BaggingClassifier(base_estimator=Rtree, n_estimators=100, random_state = 42)
```

```python
Rbagging_clf.fit(X_train,y_train)
```

```python
y_pred = Rbagging_clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

## 4. SVM

```python
from sklearn.svm import SVC
Sbagging_clf = BaggingClassifier(base_estimator=SVC(), n_estimators=100, random_state = 42)
```

```python
Sbagging_clf.fit(X_train,y_train)
```

```python
y_pred = Sbagging_clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier

ADA_clf = AdaBoostClassifier(n_estimators=100)

ADA_clf = ADA_clf.fit(X_train,y_train)

y_pred = ADA_clf.predict(X_test)
```

### 6. KNN

```python

from sklearn.neighbors import KNeighborsClassifier

KNN_clf = KNeighborsClassifier()

KNN_clf = KNN_clf.fit(X_train,y_train)
y_pred = KNN_clf.predict(X_test)
```
